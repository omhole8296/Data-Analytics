```python
td=td[td.PROD_QTY!=200]
```

```python
td['PACK_SIZE']=td['PROD_NAME'].str.extract(pat='([0-9]+g)')
```

```python
td['PACK_SIZE'] = pd.to_numeric(td['PACK_SIZE'].str.replace('g', ''), errors='coerce')
```
