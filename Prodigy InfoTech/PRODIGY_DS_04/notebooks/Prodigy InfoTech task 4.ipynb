```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(
    "twitter_training.csv",
    names=["ID", "Topic", "Sentiment", "Tweet"]
)
print(df.head())
print(df.columns)
print(df.info())
```

```python
df.dropna(inplace=True)
df["Tweet"] = df["Tweet"].astype(str)
```

```python
df["Sentiment"].value_counts().plot(kind="bar")
plt.title("Overall Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
df["Topic"].value_counts().head(10).plot(kind="bar")
plt.title("Top 10 Topics / Brands")
plt.xlabel("Topic")
plt.ylabel("Mentions")
plt.xticks(rotation=45)
plt.show()
```

```python
pd.crosstab(df["Topic"], df["Sentiment"]).head(10).plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Sentiment by Topic")
plt.xlabel("Topic")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
df["Tweet_Length"] = df["Tweet"].apply(len)

plt.hist(df["Tweet_Length"], bins=30)
plt.title("Tweet Length Distribution")
plt.xlabel("Characters")
plt.ylabel("Count")
plt.show()
```
