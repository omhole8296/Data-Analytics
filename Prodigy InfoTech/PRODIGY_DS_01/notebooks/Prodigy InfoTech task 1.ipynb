```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)
print(df.head())
print(df.columns)
```

```python
population = df["2020"].dropna()

plt.hist(population, bins=20)
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.title("Distribution of Population (2020)")
plt.show()
```

```python
top10 = df[["Country Name", "2020"]].dropna().sort_values(by="2020", ascending=False).head(10)

plt.bar(top10["Country Name"], top10["2020"])
plt.xticks(rotation=45)
plt.xlabel("Country")
plt.ylabel("Population")
plt.title("Top 10 Most Populated Countries (2020)")
plt.show()
```
